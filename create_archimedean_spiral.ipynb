```python
import numpy as np
import matplotlib.pyplot as plt
import math

# curve parameters
v = 1
c = 0

# sample parameters
ns = 500 # number of samples per spiral for total
nt = 100 # number of samples per spiral for training
dt = 0.02
t0 = 0
tf = t0 + dt*(ns-1)

# generate input
t_all = np.linspace(t0,tf,ns)

# generate CCW output
x_all_CCW = (v*t_all + c)*np.cos(t_all)
y_all_CCW = (v*t_all + c)*np.sin(t_all)

# generate CW output
x_all_CW = -(v*t_all + c)*np.cos(t_all)
y_all_CW = (v*t_all + c)*np.sin(t_all)

# visualize data
plt.plot(x_all_CCW,y_all_CCW)
plt.plot(x_all_CW,y_all_CW)
plt.show()
```
